```python
df_clean = df.copy()
```

```python
df_clean["Interest_Income"].value_counts()
```

```python
df_clean["Interest_Income"].unique()
```

```python
df_clean["Interest_Income"]=df_clean["Interest_Income"].fillna(df_clean["Interest_Income"].mean())
```

```python
df_clean["Interest_Income"].unique()
```

```python
df_clean["Interest_Income"]=df_clean["Interest_Income"].astype(int)
```

```python
df_clean["Net_Income"].value_counts()
```

```python
df_clean["Net_Income"].unique()
```

```python
df_clean["Net_Income"] = df_clean["Net_Income"].fillna(df_clean["Net_Income"].mean())
```

```python
df_clean["Net_Income"]=df_clean["Net_Income"].astype(int)
```

```python
df_clean.head()
```

```python
df_clean["Date"] = pd.to_datetime(df_clean["Date"])
```

```python
df_clean.head()
```

```python
df_clean["Year"] = pd.to_datetime(df_clean["Date"]).dt.year
df_clean["Month"] = pd.to_datetime(df_clean["Date"]).dt.month
df_clean["Day"] = pd.to_datetime(df_clean["Date"]).dt.day
```

```python
df_clean.head()
```

```python
df_clean.drop(['Date'], axis=1, inplace=True)
```
